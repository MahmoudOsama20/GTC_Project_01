```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of numerical columns to check
num_cols = ["adr", "lead_time"]

plt.figure(figsize=(12, 6))

for i, col in enumerate(num_cols, 1):
    plt.subplot(1, len(num_cols), i)  # subplots side by side
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detect for 'adr'
adr_outliers, adr_low, adr_high = detect_outliers_iqr(df, "adr")
print(f"adr outliers: {len(adr_outliers)} (below {adr_low:.2f}, above {adr_high:.2f})")

# Detect for 'lead_time'
lead_outliers, lead_low, lead_high = detect_outliers_iqr(df, "lead_time")
print(f"lead_time outliers: {len(lead_outliers)} (below {lead_low:.2f}, above {lead_high:.2f})")
```

```python
for col in num_cols:
    outliers, low, high = detect_outliers_iqr(df, col)
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[col])
    plt.axhline(low, color='red', linestyle='--')
    plt.axhline(high, color='red', linestyle='--')
    plt.title(f"Outliers in {col}")
    plt.show()
```
